# UD1 — Estadística aplicada con Python

Este notebook acompaña a la teoría de estadística aplicada de UD1. El objetivo es comprobar con código lo que se calcula a mano: media, mediana, desviación típica, IQR, outliers, correlación y gráficas.

## 1. Crear el dataset

El dataset incluye una venta extrema (`precio_total = 1000`) para analizar cómo afecta a las métricas.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'venta_id': [1, 2, 3, 4, 5, 6],
    'ciudad': ['Cádiz', 'Cádiz', 'Sevilla', 'Sevilla', 'Málaga', 'Málaga'],
    'unidades': [2, 1, 3, 4, 2, 1],
    'precio_unitario': [12, 15, 10, 11, 14, 1000],
    'precio_total': [24, 15, 30, 44, 28, 1000],
    'tiempo_entrega_h': [24, 26, 30, 28, 25, 200],
})

df
```

## 2. Medidas descriptivas

Compara la media y la mediana. Si son muy distintas, puede haber asimetría u outliers.

```python
precio = df['precio_total']

resumen = pd.Series({
    'media': precio.mean(),
    'mediana': precio.median(),
    'desviacion_tipica': precio.std(ddof=1),
    'minimo': precio.min(),
    'maximo': precio.max(),
    'rango': precio.max() - precio.min(),
})

resumen
```

## 3. Cuartiles, IQR y outliers

La regla IQR marca valores sospechosos, no valores que haya que borrar automáticamente.

```python
q1 = precio.quantile(0.25)
q3 = precio.quantile(0.75)
iqr = q3 - q1
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

outliers = df[(precio < limite_inferior) | (precio > limite_superior)]

pd.Series({
    'Q1': q1,
    'Q3': q3,
    'IQR': iqr,
    'limite_inferior': limite_inferior,
    'limite_superior': limite_superior,
})
```

```python
outliers
```

## 4. Gráficas

Las gráficas no sustituyen a las métricas. Ayudan a ver la forma de la distribución y a detectar patrones que una tabla puede ocultar.

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].hist(df['precio_total'], bins=[0, 50, 250, 500, 750, 1000], color='#4f46e5', edgecolor='white')
axes[0].set_title('Histograma')
axes[0].set_xlabel('precio_total')
axes[0].set_ylabel('frecuencia')

axes[1].boxplot(df['precio_total'], vert=False)
axes[1].set_title('Boxplot')
axes[1].set_xlabel('precio_total')

axes[2].scatter(df['precio_total'], df['tiempo_entrega_h'], color='#2563eb')
axes[2].set_title('Scatter plot')
axes[2].set_xlabel('precio_total')
axes[2].set_ylabel('tiempo_entrega_h')

plt.tight_layout()
```

## 5. Correlación

Calcula la correlación, pero interprétala con cuidado. Un único outlier puede dominar el resultado.

```python
df['precio_total'].corr(df['tiempo_entrega_h'])
```

Validamos si `precio_total` coincide con `unidades * precio_unitario`.

```python
df['precio_total_valido'] = df['precio_total'] == df['unidades'] * df['precio_unitario']
df[['venta_id', 'unidades', 'precio_unitario', 'precio_total', 'precio_total_valido']]
```
